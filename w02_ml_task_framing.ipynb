
## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

Unit of Analysis: One Row = One Unique Search Query / Content Pair.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
import numpy as np

# 1. Load dataset slice
df = pd.read_csv('../data/starter_data.csv')

# 2. Display unit of analysis (One row = One entity)
print("Data Shape:", df.shape)
display(df.head())

# 3. Create and sketch target column
df['target'] = np.where((df['ctr'] > 0.05) & (df['clicks'] > 100), 1, 0)

print("\nTarget Class Distribution:")
print(df['target'].value_counts(normalize=True))

# Display updated dataframe slice
display(df[['query_id', 'impressions', 'clicks', 'ctr', 'target']].head())
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

Non-linear Feature Interactions: Fixed rules like if CTR > 5% fail because user engagement varies wildly across search intent types, seasonality, and device types.

Adaptability: A static if/else rule requires manual updates every time search engine algorithms change, whereas an ML model continuously retrains on new data patterns without degrading over time.
