Lane: Declining Traffic Prediction

Why: I have chosen the Declining Traffic Prediction lane because identifying pages that are losing search engine visibility before they completely lose their rank is the most proactive way to protect search traffic. Predicting which content is entering a decay phase allows content teams to prioritize their rewriting and optimization efforts where they will yield the highest return on investment.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

# Bilkul sahi dataset URL se direct load karein
df = pd.read_csv('https://raw.githubusercontent.com/flyrank-bih/flyrank-ml-internship-starter/main/data/raw/content_refresh_anonymized.csv')

# Create the declining label (derived from trend_direction)
df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)

# Calculate unique pages and declining pages
# Using page_id if available, otherwise utilizing rows for unique page tracking
if 'page_id' in df.columns:
    total_pages = df['page_id'].nunique()
    declining_pages = df[df['is_declining_label'] == 1]['page_id'].nunique()
else:
    total_pages = df.shape[0]
    declining_pages = df[df['is_declining_label'] == 1].shape[0]

pct_declining = (declining_pages / total_pages) * 100

print(f"Total Pages in Dataset: {total_pages}")
print(f"Declining Pages: {declining_pages} ({pct_declining:.2f}%)")
```

The Decision: The decision this work improves is "Which specific piece of content should our writing team prioritize for a rewrite or update this week?"

The Action: A Content Strategist or SEO Specialist will take direct action by assigning these predicted-to-decay pages to writers to be refreshed with up-to-date information, new keywords, or improved structures.

The Cost of a Wrong Call:

False Positive (Wrong Alarm): If we flag a stable page as "declining", we waste valuable writer hours rewriting a healthy page that didn't need work.

False Negative (Missed Decay): If we fail to flag a page that is actually declining, we take zero action, leading to a permanent drop in rankings, lost search impressions, and decreased client revenue.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Showing the distribution of the target variable to understand the decision balance
print("Distribution of target variable (is_declining_label):")
print(df['is_declining_label'].value_counts(normalize=True))
```

Based on our initial exploratory analysis of the starter dataset, we observed these key metrics that justify dedicating 7 weeks of development to this problem:Total Sample Size: We have a solid dataset of over $1,000+$ rows to learn patterns from.Declining Rate: Around $43.9\%$ of the page entries are flagged as declining, meaning traffic decay is a very real, high-frequency threat to this website's health.Average CTR: Pages classified as declining have a significantly lower CTR on average compared to stable pages, proving that identifying these early can save real business conversions.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Real numbers from the starter data
total_rows = len(df)
decay_rate = df['is_declining_label'].mean() * 100
avg_ctr_declining = df[df['is_declining_label'] == True]['ctr'].mean()
avg_ctr_stable = df[df['is_declining_label'] == False]['ctr'].mean()

print(f"1. Total Rows in Dataset: {total_rows}")
print(f"2. General Decline Rate: {decay_rate:.2f}%")
print(f"3. Avg CTR (Declining Pages): {avg_ctr_declining:.4f}")
print(f"4. Avg CTR (Stable Pages): {avg_ctr_stable:.4f}")
```
